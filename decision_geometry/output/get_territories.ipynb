```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/vivekhsridhar/Library/Mobile Documents/com~apple~CloudDocs/Documents/Metashape/TalChhapar/output/territories_xyz.txt', header=None)
df.columns = ['idx','pos_x','pos_y','pos_z','nan']
df = df.loc[:,['idx','pos_x','pos_y']]
df.head()
```

```python
plt.scatter(df['pos_x'], df['pos_y'])
```

```python
## Z-transform territories
df['pos_x'] -= df['pos_x'].mean()
df['pos_y'] -= df['pos_y'].mean()
df['pos_x'] /= df['pos_x'].std()
df['pos_y'] /= df['pos_y'].std()

## Recentre and scale territories
df['pos_x'] *= 150
df['pos_y'] *= 150
df['pos_x'] += 500
df['pos_y'] += 500

## Convert to integers
df['pos_x'] = df['pos_x'].astype(int)
df['pos_y'] = df['pos_y'].astype(int)
```

```python
plt.scatter(df['pos_x'], df['pos_y'])
plt.xlim(0,800)
plt.ylim(100,900)
```

```python
df = df.sample(n=48)

plt.scatter(df['pos_x'], df['pos_y'])
plt.xlim(0,800)
plt.ylim(100,900)
```
