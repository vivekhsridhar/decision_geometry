```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sigma = 0.2  # Standard deviation.
k = 0.5
w = -0.5
I1 = 2.
I2 = -2.
tau = 0.5  # Time constant.
```

```python
dt = .001  # Time step.
T = 10.  # Total time.
n = int(T / dt)  # Number of time steps.
t = np.linspace(0., T, n)  # Vector of times.
```

```python
sigma_bis = sigma * np.sqrt(1. / tau)
sqrtdt = np.sqrt(dt)
```

```python
x1 = np.zeros(n)
x2 = np.zeros(n)
```

```python
for i in range(n - 1):
    x1[i + 1] = x1[i] + dt * ((I1 - k * x1[i] - w * x2[i]) / tau) + sigma_bis * sqrtdt * np.random.randn()
    x2[i + 1] = x2[i] + dt * ((I2 - k * x2[i] - w * x1[i]) / tau) + sigma_bis * sqrtdt * np.random.randn()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
ax.plot(t, x1, lw=2)
ax.plot(t, x2, lw=2)
ax.hlines(I1/(k+np.abs(w)), 0, T, linestyles='dashed')
ax.hlines(I2/(k+np.abs(w)), 0, T, linestyles='dashed')
```
